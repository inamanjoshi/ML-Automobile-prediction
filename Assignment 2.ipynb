## Symboling

Cars are initially assigned a risk factor symbol associated with its price.<br> Then, if it is more risky (or less), this symbol is adjusted by moving it up (or down) the scale.<br> Acturians call this process "symboling".<br> A value of +3 indicates that the automobile is risky, -3 that it is probably pretty safe.

## Linear regression model

```python
Lrmodel = data[['cylindernumber', 'enginesize',  'boreratio', 'horsepower','price', 'brand_category', 'mileage','fueltype',
                'aspiration', 'carbody', 'drivewheel', 'wheelbase', 'carlength', 'carwidth', 'curbweight', 'enginetype']]
```

<B> Converting categorical variable into numerical variable.

```python
cylno = pd.get_dummies(Lrmodel['cylindernumber'], drop_first = True)

Lrmodel = pd.concat([Lrmodel, cylno], axis = 1)

brand_cat = pd.get_dummies(Lrmodel['brand_category'], drop_first = True)

Lrmodel = pd.concat([Lrmodel, brand_cat], axis = 1)

eng_typ = pd.get_dummies(Lrmodel['enginetype'], drop_first = True)

Lrmodel = pd.concat([Lrmodel, eng_typ], axis = 1)

drwh = pd.get_dummies(Lrmodel['drivewheel'], drop_first = True)

Lrmodel = pd.concat([Lrmodel, drwh], axis = 1)

carbo = pd.get_dummies(Lrmodel['carbody'], drop_first = True)

Lrmodel = pd.concat([Lrmodel, carbo], axis = 1)

aspir = pd.get_dummies(Lrmodel['aspiration'], drop_first = True)

Lrmodel = pd.concat([Lrmodel, aspir], axis = 1)

fuelty = pd.get_dummies(Lrmodel['fueltype'], drop_first = True)

Lrmodel = pd.concat([Lrmodel, fuelty], axis = 1)

Lrmodel.drop(['fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginetype', 'cylindernumber','brand_category'], axis = 1, inplace = True)
```

## Splitting the Data into Training and Testing sets

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)

df_train, df_test = train_test_split(Lrmodel, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
num_vars = ['enginesize','boreratio', 'horsepower', 'price','mileage','wheelbase', 'carlength', 'carwidth', 'curbweight']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_train.head()
```

```python
df_train.describe()
```

### Correlation

```python
plt.figure(figsize = (16, 10))
sns.heatmap(df_train.corr(), annot = True)
plt.show()
```

<B> Getting the x,y train set for model building.

```python
X_train = df_train
y_train = df_train.pop('price')
```

<B> Recursive feature elimination

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 10)       
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

<B> Building model using statsmodel

```python
X_train_rfe = X_train[col]
```

```python
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()
```

```python
print(lm.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
X_train_new1 = X_train_rfe.drop(["twelve"], axis = 1)


# Adding a constant variable 
import statsmodels.api as sm  
X_train_lm = sm.add_constant(X_train_new1)

lm = sm.OLS(y_train,X_train_lm).fit()  

#Let's see the summary of our linear model
print(lm.summary())
```
